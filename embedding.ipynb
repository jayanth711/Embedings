```python
embedings_1024= OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embedings_1024
```

```python
text="This is tutorial on OpenAi embedding"
query_result = embedings_1024.embed_query(text)
len(query_result)
```

```python
### TextLoader

from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter= RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc = text_splitter.split_documents(docs)
final_doc
```

```python
## vector embedding and vector storeDB

from langchain_community.vectorstores import Chroma

db= Chroma.from_documents(final_doc, embedings_1024)
db
```

```python
### Retrive the results from query vectorstore DB
query= "involves the systematic collection and preparation of data from various sources, ensuring that our models"
retrive_result =db.similarity_search(query)
print(retrive_result)
```
